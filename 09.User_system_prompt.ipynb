Creating bounding boxes with unique label

```python
import cv2

# Function to draw multiple bounding boxes with labels and numbers
def draw_bounding_boxes(image_path, boxes, output_path):

    # Read the image
    image = cv2.imread(image_path)

    if image is None:
        print("Error: Could not open or find the image.")
        return

    # Define the color (BGR) and thickness of the bounding box
    box_color = (0, 255, 0)  # Green color
    box_thickness = 2

    # Define the font and scale for the label
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.3
    font_color = (255, 0, 255)  # Red color
    font_thickness = 1

    # Loop through each bounding box and label
    for i, (x1, y1, x2, y2) in enumerate(boxes):
        # Convert coordinates to integers
        x1, y1, x2, y2 = map(int, (x1, y1, x2, y2))

        # Draw the bounding box
        cv2.rectangle(image, (x1, y1), (x2, y2), box_color, box_thickness)

        # Prepare the label with the number
        label_with_number = f"{i+1}"

        # Calculate the position for the label (just above the bounding box)
        label_size = cv2.getTextSize(label_with_number, font, font_scale, font_thickness)[0]
        label_x = x1 + 5
        label_y = y1 + label_size[1] + 5

        # Draw the label background
        cv2.rectangle(image, (label_x - 5, label_y - label_size[1] - 5),
                      (label_x + label_size[0] + 5, label_y + 5), box_color, -1)

        # Put the label on the image
        cv2.putText(image, label_with_number, (label_x, label_y), font, font_scale, font_color, font_thickness)

    # Save the output image
    cv2.imwrite(output_path, image)
    print(f"Output image saved at {output_path}")

```
